# 02 Feature Extraction

This notebook inspects the transparent features used by the detector.


## Current feature set

- `contains_link`
- `contains_password`
- `contains_urgent`
- `contains_reward`
- `len_message`


```python
import pandas as pd
from pathlib import Path
import json

DATA_DIR = Path("/workspace/data")
ingested_path = DATA_DIR / "ingested_events.jsonl"

rows = []
if ingested_path.exists():
    with open(ingested_path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if line:
                rows.append(json.loads(line))

print("Loaded rows:", len(rows))
```

```python
feature_rows = []
for r in rows:
    anon = r.get("anon_record", {})
    features = r.get("features", {})
    dr = r.get("detector_result", {})

    feature_rows.append({
        "source": anon.get("source"),
        "event_type": anon.get("event_type"),
        "language": anon.get("language"),
        "contains_link": features.get("contains_link"),
        "contains_password": features.get("contains_password"),
        "contains_urgent": features.get("contains_urgent"),
        "contains_reward": features.get("contains_reward"),
        "len_message": features.get("len_message"),
        "label": dr.get("label"),
        "action": dr.get("action"),
    })

df = pd.DataFrame(feature_rows)
df.head()
```

```python
if not df.empty:
    display(df[[
        "contains_link",
        "contains_password",
        "contains_urgent",
        "contains_reward",
        "len_message"
    ]].describe(include="all"))
else:
    print("No ingested events found yet.")
```

```python
if not df.empty:
    print("Feature activation totals")
    display(df[["contains_link", "contains_password", "contains_urgent", "contains_reward"]].sum())
```

## Interpretation

These features are intentionally simple so that students and reviewers can understand why a message received a given score.
